```python
# Nỗi đau (Pain Point)
Thử tưởng tượng cái App của em là một ngôi nhà, mà cửa thì không khóa, ai thích thì vào. Xong vào trong rồi, ông khách lại đi lục tung ngăn kéo của chủ nhà, thậm chí xóa luôn cả sổ hộ khẩu.
- **Không có Authentication**: Em không biết ai là ai.
- **Không có Authorization**: Em không kiểm soát được ai được làm gì. Kết quả là hệ thống loạn cào cào, dữ liệu "bay màu" lúc nào không hay.

---

# Giải pháp (Solution)
Chúng ta sinh ra bộ đôi này để thiết lập trật tự:
- **Authentication**: Để biết "Ông là ai?" (Xác nhận danh tính).
- **Authorization**: Để biết "Ông có quyền làm việc này không?" (Kiểm soát truy cập).

---

# Khái niệm & Tại sao (ELI5)
- **Authentication (Xác thực)**: Là bước đầu tiên. Em đưa thẻ căn cước hoặc nhập Username/Password. Nếu đúng, hệ thống nói: "Ok, đúng là em rồi!".
- **Authorization (Phân quyền)**: Sau khi vào được rồi, hệ thống check xem em là "User" hay "Admin". User thì chỉ được xem bài viết, Admin mới được xóa bài.

---

# Ví dụ Liên Quân Mobile
- **Authentication**: Là lúc em mở app và đăng nhập tài khoản Garena.
	- Nếu sai mật khẩu -> Hệ thống đuổi ra (401 Unauthorized - thực chất là Unauthenticated).
	- Nếu đúng -> Hệ thống xác nhận: "Chào mừng tuyển thủ ABC đã quay trở lại".

- **Authorization**: Sau khi vào sảnh chờ, em bấm vào "Đấu Hạng".
	- Nếu em chưa đủ cấp 6 hoặc chưa đủ 5 tướng -> Hệ thống báo: "Bạn chưa đủ điều kiện tham gia".
	- Đó chính là Phân quyền! Em đã đăng nhập thành công (AuthN ok), nhưng em chưa đủ "quyền" để chơi rank (AuthZ fail).
```

```python
# Tóm lại
- Bước xác minh họ là ai/xác minh danh tính (cần credentials): Authentication
- Bước đảm bảo được phép mới đc truy cập: Authorization
```

```python
1. Bước đăng nhập (Mang theo thông tin xác thực) --> token --> Authentication (Xác minh danh tính)
2. Cầm token đó truy cập vào tài nguyên được bảo vệ --> Authorization (Phân quyền)
```

```python
# Authentication
- Cầm THÔNG TIN ĐĂNG NHẬP (`credentials` - username/password) để tạo ra token/session

# Authorization
- Kiểm tra quyền hạn

# Hiểu lầm
- Nếu truy cập vào tài nguyên được bảo vệ. VD: `/api/users` chia  thành 2 bước:
	1. Bắn token, và Backend kiểm tra token hợp lệ. Nếu hợp lệ thì cho truy cập vào tài nguyên được bảo vệ
	> Hiểu lầm: Bước kiểm tra token hợp lệ --> người ta xếp vào `Authentication`
	2. Bước lấy dữ liệu ra --> check quyền hạn xem người dùng có quyền hay không --> thì là `Authorization`
> Hiểu lầm phổ biến!!!

# Đính chính
- Khi tôi cầm thông tin đăng nhập (email, password) bắn lên để tôi xác nhận đó là người dùng nào
	> Đây là bước `Authentication`
	> Sau đó tôi nhận về session/token
- Sau đó tôi truy cập vào tài nguyên được bảo vệ
	> Kiểm tra token (Tuy nhiên token này không được coi là `credentials`). Vậy nên nó không được coi là `Authentication`
	> Nó xác minh token này hợp lệ không thì vẫn nó thuộc bước kiểm tra phân quyền --> `Author`. Token hợp lệ thì mày có quyền để vào (Chúng ta hiện bỏ qua Phân quyền)

	> Tóm lại: kiểm tra tính hợp lệ của token (xảy ra sau khi `authentication`) là bước `authorization`
	> Tính năng phân quyền cũng là `authorization`
```

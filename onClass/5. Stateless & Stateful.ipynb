```python
Để hiểu về Stateless và Stateful, anh em mình không nên học thuộc lòng định nghĩa khô khan. Hãy nhìn nó dưới góc độ: **"Làm sao để Server nhận ra mình là ai?"**
---
# Giải thích Khái niệm Stateless & Stateful
- Nỗi đau (Pain Point)
- Giao thức HTTP mà anh em mình dùng hàng ngày nó có một đặc tính rất "phũ": 
	1. Nó bị mất trí nhớ (Stateless). 
	2. Cứ mỗi khi em gửi một Request lên Server, nó coi em như người lạ qua đường. Em vừa đăng nhập xong ở Request 1, sang Request 2 hỏi "Lấy danh sách bài viết của tôi", Server sẽ hỏi ngược lại: "Ơ, bạn là ai? Tôi không biết, bạn chưa đưa chứng minh thư".

# Giải pháp (Solution)
- Để giải quyết cái "bệnh mất trí" này, chúng ta có 2 trường phái:
	+ **Stateful** (Lưu trạng thái): Server tự ghi chép vào một cuốn sổ (Session/Database) để nhớ mặt khách.
	+ **Stateless** (Không lưu trạng thái): Server không thèm nhớ gì cả, khách tự mang theo một cái "thẻ bài" (Token) có đầy đủ thông tin để Server tự kiểm tra tại chỗ.

# Khái niệm & Tại sao (ELI5)
- **Stateful** (Session-based): 
	+ Giống như em đi gửi xe. 
	+ bác bảo vệ phải giữ một "cuốn sổ" hoặc một hệ thống máy tính để ghi nhớ: "Số thẻ 123 này là của ông Sơn, đi xe SH màu đen".
	+ Vấn đề 1: Nếu bãi xe quá đông (hàng triệu xe), bác bảo vệ phải có một cuốn sổ khổng lồ, tìm kiếm rất lâu (Server tốn RAM/Database).
	+ Vấn đề 2: Nếu em sang gửi ở bãi xe cơ sở 2 của cùng hệ thống, bác bảo vệ bên đó không có cuốn sổ của cơ sở 1, nên bác ấy không nhận ra em (Khó mở rộng hệ thống sang nhiều Server).

- **Stateless** (Token-based/JWT): 
	+ Bây giờ, hãy tưởng tượng một bãi xe "thông minh" kiểu Stateless. Bác bảo vệ không dùng sổ, cũng không giữ xe của em trong danh sách theo dõi.
	+ Thay vào đó, quy trình như sau:
	1. Đăng ký: Em đến gặp bác bảo vệ lần đầu, đưa căn cước công dân.
	2. **Cấp "Thẻ Thông Minh"**: Bác ấy in cho em một cái thẻ nhựa, trên đó ghi rõ: "Ông Sơn - Xe SH đen - Biển số 29-F1 - Được gửi đến 22h đêm". Sau đó, bác ấy dùng một cái con dấu chống giả (Signature) đóng "cộp" một phát lên thẻ.
	3. **Sử dụng**: Mỗi lần em vào bãi, em chỉ cần giơ cái thẻ đó ra. Bác bảo vệ chỉ cần nhìn con dấu xem có đúng là dấu của mình không. Nếu đúng, bác ấy đọc thông tin trên thẻ là biết ngay em là ai, xe gì. Bác ấy không cần tra sổ vì mọi thông tin đã nằm trên tay em rồi.

# Tóm lại:
- Stateless nghĩa là Server "mất trí nhớ" tự nguyện. Nó không thèm nhớ em là ai sau khi em rời đi.
- Nó đẩy trách nhiệm "ghi nhớ" sang cho em (Client). Em phải tự giữ cái "thẻ thông minh" (JWT) đó và trình ra mỗi khi cần.
```
